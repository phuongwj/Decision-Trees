```python
# Importing libraries and collections

import pandas as pd
import numpy as np
from collections import Counter
from sklearn.model_selection import train_test_split
from pprint import pprint
```

**Loading the Titanic dataset and removing any rows with missing data in the three features - Age, Sex, Pclass**

```python
# loading the Titanic dataset into a DataFrame
df = pd.read_csv('train.csv')

# Only keeping necessary columns
df = df[['Age', 'Sex', 'Pclass', 'Survived']]

# Features
features = ['Age', 'Sex', 'Pclass']

# Dropping rows with missing data in the three features
df = df.dropna(subset=features)
```

**Feature Engineering for handling Age**

```python
def age_handle(age):

  if age <= 12:
    return 'Child'
  elif age <= 19:
    return 'Teen'
  elif age <= 64:
    return 'Adult'
  else:
    return 'Elderly'

df['Age_Category'] = df['Age'].apply(age_handle)

df = df.drop('Age', axis = 1)

features = ['Sex', 'Pclass', 'Age_Category']

target = df['Survived']
```

**Function to calculate Entropy of Target Variable (i.e. Survived)**

```python
def entropy(target_values):
  """
  Calculates the Entropy of the target variable - Survived

  Args:
    target_values: List of target values from Survived

  Returns:
    Float representing the entropy
  """

   # Total number of samples
  total = len(target_values)

  # Count occurrences of 1s and 0s (survived and didn't survive)
  count_survived = sum(target_values)
  count_died = total - count_survived

  # Probability distribution of survived vs didn't survive
  probability_survived = count_survived / total
  probability_died = count_died / total

  enpy = 0

  # Applying Entropy formula
  if probability_survived > 0:
    enpy -= probability_survived * np.log2(probability_survived)

  if probability_died > 0:
    enpy -= probability_died * np.log2(probability_died)

  return enpy
```

**Function to calculate Information Gain of a Feature**

```python
def information_gain(data, feature, target):
  """
  Calculates information gain for a specified feature

  Args:
    data: The DataFrame with all the data
    feature: The column we're testing (either 'Sex', 'Pclass', or 'Age')
    target: 'Survived' target, what we're predicting

  Returns:
    Float representing the information gain of a feature
  """

  # Total entropy
  enpy_before = entropy(data[target])

  # Weighted entropy after split
  unique_values = data[feature].unique()
  total_rows = len(data)
  weighted_enpy_after = 0

  for value in unique_values:
    condition = data[feature] == value
    subset = data[condition]

    # Calculating the specific conditional Entropy
    probability_of_value = len(subset) / total_rows
    enpy_of_subset = entropy(subset[target])

    # Calculating the conditional information gain
    weighted_enpy_after += probability_of_value * enpy_of_subset

  # Information gain calculation
  info_gain = enpy_before - weighted_enpy_after
  return info_gain
```

**Function to find the feature with the largest Information Gain that can be used for splitting**

```python
def find_best_split(data, list_of_features, target):
  """
  Finds the feature with the highest information gain

  Args:
    data: The DataFrame with all the data
    list_of_features: List of all the features (either 'Sex', 'Pclass', or 'Age')
    target: 'Survived'

  Returns:
    A tuple with the best feature name and their info gain
  """

  # Variables initialization
  best_feature_name = None
  highest_info_gain = -1

  # Finding the feature with the highest information gain
  for feature_name in list_of_features:

    # Calculating the information gain of a specified feature
    curr_gain = information_gain(data, feature_name, target)

    if curr_gain > highest_info_gain:
      highest_info_gain = curr_gain
      best_feature_name = feature_name

  return best_feature_name, highest_info_gain

```

**Helper function used in the decision tree when we hit a stopping condition**

```python
def majority_class(target_values):
  """
  Returns the most frequent value

  Arg:
    target_values: List of target values (0s and 1s for Survived)

  Returns:
    int: The most common value (0 or 1)
  """

  counts = Counter(target_values)
  return max(counts, key=counts.get)
```

**Recursive function to build the decision tree**

```python
def decision_tree(data, features, target):
  """
  Recursively builds a decision tree

  Args:
    data: The DataFrame with all the data
    features: List of feature column names
    target: Target column name (i.e. Survived)

  Return:
    Dictionary representing the tree
  """

  # Base case 1, check if all samples have the same label
  if len(data[target].unique()) == 1:
    return data[target].iloc[0] # Here, iloc is used for accessing rows and columns in a DataFrame

  # Base case 2, check if we have features left
  if len(features) == 0:
    return majority_class(data[target])

  # Finds the best feature to split on
  best_feature, best_info_gain = find_best_split(data, features, target)

  # If no information gain, return majority class
  if best_info_gain == 0:
    return majority_class(data[target])

  # Create a tree node
  tree = {best_feature: {}}

  # Removes a feature from remaining features
  remaining_features = []
  for feature in features:
    if feature != best_feature:
      remaining_features.append(feature)

  # Recursively builds subtrees for each value of the best feature
  for value in data[best_feature].unique():
    condition = data[best_feature] == value
    subset = data[condition]

    if len(subset) == 0:
      tree[best_feature][value] = majority_class(data[target])
    else:
      tree[best_feature][value] = decision_tree(subset, remaining_features, target)

  return tree
```

**Predict functions**

```python
def predict_one_person(tree, person):
  """
  Walk down the decision tree to predict for one person

  Args:
    tree: Decision tree (dictionary) or final prediction (0 or 1)
    person: One row from DataFrame

  Returns:
    0 (died) or 1 (survived)
  """

  # Check if we have reached the bottom of the tree (i.e. 0 or 1)
  if not isinstance(tree, dict):
    return tree

  # Find our what feature this node asks about
  feature_to_check = list(tree.keys())[0]

  # Get this person's avlue for that feature
  persons_value = person[feature_to_check]

  # Check if this value exists in our tree
  if persons_value not in tree[feature_to_check]:
    return 0

  next_level_of_tree = tree[feature_to_check][persons_value]
  return predict_one_person(next_level_of_tree, person)
```

```python
def predict_all_people(tree, test_data):
  """
  Predict for all people in the test set

  Args:
    tree: Decision tree
    test_data: DataFrame with many rows

  Returns:
    List of predictions
  """

  predictions = []
  number_of_people = len(test_data)

  for i in range(number_of_people):

    # Get one person's data (i.e. 1 row of data)
    one_person = test_data.iloc[i]

    # Make a prediction for them
    their_prediction = predict_one_person(tree, one_person)

    # Save the prediction
    predictions.append(their_prediction)

  return predictions
```

**5 Train-Test Splits**

```python
test_sizes = [0.5, 0.4, 0.3, 0.2, 0.1]

results = []

for test_size in test_sizes:
  train_size = 1.0 - test_size

  # Splitting the data into training and testing sets (much like Assignment 1)
  training_set, testing_set, training_target, testing_target = train_test_split(
      df, target, test_size=test_size
  )

  # Building the tree
  training_data = training_set.copy()
  training_data['Survived'] = training_target

  tree = decision_tree(training_data, features, 'Survived')

  # Predict on test set
  predictions = predict_all_people(tree, testing_set)

  # Calculate accuracy
  correct_predictions = 0
  total_predictions = len(testing_target)

  for i in range(len(predictions)):
    actual = testing_target.iloc[i]
    predicted = predictions[i]

    if actual == predicted:
      correct_predictions += 1

  accuracy = correct_predictions / total_predictions

  # Store results
  results.append({
      'train_size': train_size,
      'test_size': test_size,
      'accuracy': accuracy
  })

print("\nSummary: ")
pprint(results)
```
